```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
books = pd.read_csv('dataset/books.csv')
book_tags = pd.read_csv('dataset/book_tags.csv')
tags = pd.read_csv('dataset/tags.csv')
```

```python
tags.count()
```

```python
book_tags.count()
```

```python
books.count()
```

```python
data_counts = {
    'book_tags.csv': book_tags.shape[0],
    'books.csv': books.shape[0],
    'tags.csv': tags.shape[0]
}
```

```python
plt.figure(figsize=(10, 6))
barplot = sns.barplot(x=list(data_counts.keys()), y=list(data_counts.values()), palette='pastel')
plt.xlabel('Dosya Adı', fontsize=18)
plt.ylabel('Satır Sayısı', fontsize=18)
plt.title('Her Dosyadaki Veri Sayısı (GoodBooks-10k)')
plt.xticks(rotation=45, fontsize=20)
plt.yticks(fontsize=20)

for index, value in enumerate(data_counts.values()):
    barplot.text(index, value + 10000, f'{value:,}', ha='center', va='bottom', fontsize=18)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.set(font_scale=1.2)  # Seaborn'un varsayılan font boyutunu 1.2 katına çıkarıyoruz

barplot = sns.barplot(x=list(data_counts.keys()), y=list(data_counts.values()), palette='pastel')
plt.xlabel('Dosya Adı', fontsize=18)
plt.ylabel('Satır Sayısı', fontsize=18)
plt.title('Her Dosyadaki Veri Sayısı (GoodBooks-10k)', fontsize=18)
plt.xticks(rotation=45, fontsize=20)
plt.yticks(fontsize=20)

for index, value in enumerate(data_counts.values()):
    barplot.text(index, value + 10000, f'{value:,}', ha='center', va='bottom', fontsize=18)

plt.tight_layout()
plt.show()
```
